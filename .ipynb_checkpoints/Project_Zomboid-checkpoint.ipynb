# 4.  Insights

## 4.1 ข้อสมมติฐานว่าเขตบนเกาะหรือส่วน Vashon Island กับ Mercer Island นั้นจะมีในบ้านที่มีสภาพและคุณภาพใกล้เคียงกัน จะมีราคาที่ถูกกว่า
จากแผนที่เราจะทราบว่าเขต King County , USA นั้นจะมีพื้นที่บนบกและที่เป็นพื้นที่บนเกาะอยู่โดยจะมีเกาะที่ชื่อว่า Vashon Island กับ Mercer Island
ข้อสมมติฐานว่าเขตบนเกาะหรือส่วน Vashon Island กับ Mercer Island นั้นจะมีในบ้านที่มีสภาพและคุณภาพใกล้เคียงกันจะมีราคาที่ถูกกว่า

## ตารางความแตกต่างระหว่างบนเกาะกับบนบก

```python
kc_house_data_copy = kc_house_data.copy()
kc_house_data_copy["Group"] = 0
kc_house_data_copy.loc[(kc_house_data_copy.Grade >= 1) & (kc_house_data_copy.Grade <=5) , "Group"] = "Bad"
kc_house_data_copy.loc[(kc_house_data_copy.Grade >= 6) & (kc_house_data_copy.Grade <=9) , "Group"] = "Average"
kc_house_data_copy.loc[(kc_house_data_copy.Grade >= 10) & (kc_house_data_copy.Grade <=13) , "Group"] = "Good"

kc_house_data_copy["Area"] = "On Land"

kc_house_data_copy.loc[(kc_house_data.Zipcode == 98013) | (kc_house_data.Zipcode == 98070) | (kc_house_data.Zipcode == 98040), "Area"] = "Island"

kc_house_data_copy = kc_house_data_copy.drop('ID', axis = 1)
kc_house_data_copy = kc_house_data_copy.drop('Zipcode', axis = 1)
kc_house_data_copy = kc_house_data_copy.drop('Latitude', axis = 1)
kc_house_data_copy = kc_house_data_copy.drop('Longtitude', axis = 1)
kc_house_data_copy = kc_house_data_copy.drop('yr_built', axis = 1)
kc_house_data_copy = kc_house_data_copy.drop('yr_renovated', axis = 1)
kc_house_data_copy = kc_house_data_copy.drop('Date', axis = 1)

kc_house_data_copy['Group'] = pd.Categorical(kc_house_data_copy['Group'], ["Bad", "Average", "Good"])
kc_house_data_copy = kc_house_data_copy.set_index(['Area', 'Group'], inplace=False)
kc_house_data_copy = kc_house_data_copy.groupby(["Area","Group"]).mean()
kc_house_data_copy.loc[:, "Price"] = kc_house_data_copy["Price"].map('{:,.3f}'.format)

kc_house_data_copy
```

## ความแตกต่างทางด้านราคาระหว่างบนบกกับบนเกาะ

```python
bad_price_dif = float((kc_house_data_copy.loc[('On Land','Bad'),:].Price).replace(",", "")) / float((kc_house_data_copy.loc[('Island','Bad'),:].Price).replace(",", "")) * 100
average_price_dif = float((kc_house_data_copy.loc[('On Land','Average'),:].Price).replace(",", "")) / float((kc_house_data_copy.loc[('Island','Average'),:].Price).replace(",", "")) * 100
good_price_dif =  float((kc_house_data_copy.loc[('On Land','Good'),:].Price).replace(",", "")) / float((kc_house_data_copy.loc[('Island','Good'),:].Price).replace(",", "")) * 100

    
print(" On Land and Island diffence of price in percentage in Bad Group is " + str(math.floor(bad_price_dif))+ "%\n",
      "On Land and Island diffence of price in percentage in Average Group is " + str(math.floor(average_price_dif))+ "%\n",
      "On Land and Island diffence of price in percentage in Good Group is " + str(math.floor(good_price_dif))+ "%")



```

- จากสมมุติฐานจะพบได้ว่าได้ผลสรุปว่าผิด เนื่องจากราคาโดยเฉลี่ยของบนเกาะและบนบกนั้นมีราคาต่างกันอย่างมาก โดยเราจะพบว่าบนเกาะนั้นจะมีราคาโดยเฉลี่ยที่สูงกว่ามาก 

- โดยเราแบ่งกลุ่มบ้านแยกเป็นสามส่วนคือ Bad, Average กับ Good โดยใช้เกณฑ์แบ่งโดยดูจาก Grade ของบ้าน หากมี Grade ตั้งแต่ 1-5 จะ จัดอยู่ในกลุ่ม Bad หากอยู่ระหว่าง 6-9 จะจัดอยู่ในกลุ่ม Average และถ้าอยู่ระหว่าง 10-13 จะจัดอยู่ในกลุ่ม Good

- เรายังพบต่ออีกว่าราคาเฉลี่ยของแต่ละกลุ่มนั้น บนเกาะจะมีราคาสูงกว่าประมาณตั้งแต่ 61 เปอร์เซ็น ถึง 76 เปอร์เซ็นโดยเฉลี่ย แต่ในขณะเดียวกันบ้านโดยเฉลี่ยบนเกาะนั้นจะให้ฟังค์ชั่นได้น้อยกว่า กล่าวคือโดยเฉลี่ยนั้นจะมีห้องนอนที่น้อยกว่า และ ห้องน้ำที่น้อยกว่าอีกด้วย

## 4.2 ราคาบ้านหลังจากการ Renovate แล้วราคาจะถูกขึ้นหรือแพงขึ้น

## 4.3 เปอร์เซ็นความเปลี่ยนแปลงของราคาบ้านที่ขายไปตั้งแต่เดือนพฤษภาคม ปี 2014 ถึง เดือน พฤษภาคม ปี 2015

```python
kc_house_data_copy = kc_house_data.copy()
kc_house_data_copy['Date'] = pd.to_datetime(kc_house_data_copy['Date'])

kc_house_data_copy.set_index('Date', inplace=True)
kc_house_data_copy = kc_house_data_copy.drop("ID",axis = 1)
kc_house_data_copy = kc_house_data_copy.drop("Zipcode",axis = 1)
kc_house_data_copy = kc_house_data_copy.drop("Latitude",axis = 1)
kc_house_data_copy = kc_house_data_copy.drop("Longtitude",axis = 1)
kc_house_data_copy = kc_house_data_copy.drop("yr_renovated",axis = 1)


kc_house_data_copy_pct = kc_house_data_copy.copy()



kc_house_data_copy_pct = kc_house_data_copy_pct.groupby(pd.Grouper(freq='M')).mean()
kc_house_data_copy_pct['Pct_Price_Change'] = (kc_house_data_copy_pct['Price'].pct_change().fillna(0) + 1).cumprod() * 100 - 100
kc_house_data_copy_pct
```

## กราฟราคาบ้านเฉลี่ยในแต่ละเดือน ภายในปี 2014-2015

```python
sb.lineplot(data=kc_house_data_copy_pct, x=kc_house_data_copy_pct.index, y=kc_house_data_copy_pct.Pct_Price_Change, sizes = 10 )
```

-  เราได้ทำการสร้าง Column ใหม่ที่ชื่อว่า 'Pct Price Change' โดยที่ Column นี้จะจัดเก็บเปอร์เซ็นความเปลี่ยนแปลงของราคาโดยนำทุกค่าความเฉลี่ยของทุกๆเดือนในข้อมูล (เดือน พฤษภาคม ของปี 2014 ถึง เดือน พฤษภาคม ของปี 2015) มา แล้วนำมาเทียบกับราคาบ้านโดยเฉลี่ยของเดือน พฤษภาคม ปี 2014 เพื่อดูว่าตั้งแต่เริ่มต้นจนสิ้นสุดของข้อมูลนั้น ราคาบ้านได้มีการเปลี่ยนแปลงหรือเปล่า แล้วเราสามารถหาข้อสรุปอะไรได้บ้างหากมีการเปลี่ยนแปลง

- จากกราฟเราจะพบว่าเปอร์เซ็นความเปลี่ยนแปลงของบ้านตั้งแต่เดือน พฤษภาคม ของปี 2014 นั้นจะพบว่าราคาขายบ้านได้ค่อยๆลดลงไปเรื่อยๆ จนกระทั่งก่อนถึงเดือน พฤศจิกายน ที่เปอร์เซ็นราคาความเปลี่ยนแปลงนั้นได้พยายามที่จะกู้กลับคืนมา แต่พอเข้าเดือน พฤศจิกายน ก็กลับลดลงต่ำไปกว่าเดิมยาวไปจนถึงเดือนเดือนกุมภาพันธ์ของปี 2015 โดยมีเดือนกุมภาพันธ์นั้นเองที่มีราคาบ้านน้อยที่สุด

- แต่เมื่อเข้าเดือนมีนาคมมานั้นเอง ราคาก็ได้ดีดกลับมาในราคาปกติเหมือนเดือน พฤษภาคม ปี 2014 แล้วยังพบอีกว่าราคาบ้านในเดือน เมษายน ปี 2015 และ เดือน พฤษภาคม ปี 2015 นั้นได้มีราคาบ้านโดยเฉลี่ยที่สูงสุดอีกด้วย

```python
kc_house_data_season = kc_house_data_copy.copy()
kc_house_data_season["Season"] = 0
kc_house_data_season.loc[(kc_house_data_season.index.month >= 3) & (kc_house_data_season.index.month <= 5) & (kc_house_data_season.index.year == 2014), "Season"] = "2014 Spring"
kc_house_data_season.loc[(kc_house_data_season.index.month >= 6) & (kc_house_data_season.index.month <= 8) & (kc_house_data_season.index.year == 2014), "Season"] = "2014 Summer"
kc_house_data_season.loc[(kc_house_data_season.index.month >= 9) & (kc_house_data_season.index.month <= 11) & (kc_house_data_season.index.year == 2014), "Season"] = "2014 Fall"
kc_house_data_season.loc[(kc_house_data_season.index.month >= 12) | (kc_house_data_season.index.month <= 2), "Season"] = "2014-2015 Winter"

kc_house_data_season.loc[(kc_house_data_season.index.month >= 3) & (kc_house_data_season.index.month <= 5) & (kc_house_data_season.index.year == 2015), "Season"] = "2015 Spring"


kc_house_data_season.set_index('Season', inplace=True)
kc_house_data_season = kc_house_data_season.groupby('Season').mean()

season_order = ["2014 Spring", "2014 Summer", "2014 Fall", "2014-2015 Winter", "2015 Spring"]

kc_house_data_season = kc_house_data_season.reindex(season_order)
kc_house_data_season = kc_house_data_season.dropna()
kc_house_data_season['Pct_Price_Change'] = (kc_house_data_season['Price'].pct_change().fillna(0) + 1).cumprod() * 100 - 100


kc_house_data_season = kc_house_data_season.drop('sqft_above', axis = 1)
kc_house_data_season = kc_house_data_season.drop('sqft_basement', axis = 1)
kc_house_data_season
```

```python
sb.lineplot(data=kc_house_data_season, x=kc_house_data_season.index, y=kc_house_data_season.Pct_Price_Change, sizes = 10 )
```

- เมื่อดูจากกราฟโดยการดูจากฤดูแทน ผลสรุปก็จะชี้ชัดว่าราคาเฉลี่ยของบ้านนั้นก็ยังขึ้นอยู่กับฤดู โดยพบราคาจะอยู่จุดตกต่ำสุดในฤดูหนาว และจะเพิ่มสูงขึ้นในฤดูใบไม้ผลิหรือก่อนฤดูร้อน

- และเนื่องจากว่าราคาบ้านโดยดูจากเปอร์เซ็นความเปลี่ยนแปลงของราคานั้นสูงขึ้นในปี 2015 ยังอาจเป็นข้อสรุปได้อีกว่าราคาบ้านโดยเฉลี่ยของ King County นั้นอาจจะมีราคาที่สูงเพิ่มขึ้นไปเรื่อยๆทุกปีอีกด้วย
